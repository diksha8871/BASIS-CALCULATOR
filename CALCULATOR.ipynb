```python
import tkinter as tk

# Function to append button text to the input field
def on_click(value):
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(0, current + value)

# Function to clear the entry field
def clear():
    entry.delete(0, tk.END)

# Function to evaluate the expression
def calculate():
    try:
        result = eval(entry.get())
        entry.delete(0, tk.END)
        entry.insert(0, str(result))
    except Exception:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

# Create main window
root = tk.Tk()
root.title("Basic Calculator")
root.configure(bg="black")

# Entry field
entry = tk.Entry(root, font=("Arial", 24), bd=5, relief=tk.RIDGE, bg="black", fg="white", justify="right")
entry.grid(row=0, column=0, columnspan=4, padx=10, pady=20, ipadx=8, ipady=8)

# Button configuration
button_config = {
    "font": ("Arial", 18),
    "bd": 2,
    "relief": tk.RIDGE,
    "width": 4,
    "height": 2
}

# Button layout
buttons = [
    ('7', '8', '9', '/'),
    ('4', '5', '6', '*'),
    ('1', '2', '3', '-'),
    ('0', 'C', '=', '+')
]

# Create and place buttons
for row_index, row in enumerate(buttons):
    for col_index, symbol in enumerate(row):
        if symbol.isdigit():  # Numeric buttons: white background
            bg_color = "white"
            fg_color = "black"
        elif symbol == 'C':  # Clear button: red background
            bg_color = "red"
            fg_color = "white"
        else:  # Arithmetic buttons: blue background
            bg_color = "skyblue"
            fg_color = "black"

        btn = tk.Button(root, text=symbol, bg=bg_color, fg=fg_color, **button_config)

        # Assign functions
        if symbol == '=':
            btn.config(command=calculate)
        elif symbol == 'C':
            btn.config(command=clear)
        else:
            btn.config(command=lambda val=symbol: on_click(val))

        btn.grid(row=row_index + 1, column=col_index, padx=5, pady=5)

# Run the application
root.mainloop()
```
